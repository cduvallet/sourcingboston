This notebook parses latitude and longitude

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/finalDraft.csv', index_col=0)
df.head()
```

```python
df['latitude'] = df['Location'].str.strip('(').str.split(',').str[0]
df['longitude'] = df['Location'].str.strip(')').str.split(',').str[1]
df.head()
```

```python
df.to_csv('../data/finalDraft_with_lat_long.csv', index=False)
```
